```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Post_ID': range(1, 21),
    'Platform': ['Instagram']*10 + ['YouTube']*10,
    'Likes': [120,230,340,560,420,380,290,610,500,450,
              200,310,400,520,470,360,280,490,430,390],
    'Comments': [12,25,30,55,40,35,28,60,50,45,
                 20,30,38,45,42,33,25,40,37,34],
    'Shares': [8,15,20,30,25,22,18,35,28,26,
               10,18,22,27,24,20,15,25,23,21],
    'Impressions': [2000,3000,4000,6000,5000,4500,3500,6500,5800,5200,
                     2500,3200,4100,5500,5100,4300,3300,4800,4600,4200],
    'Posting_Hour': [9,11,13,18,20,19,17,21,22,16,
                     10,12,14,18,19,20,16,21,22,15]
}

df = pd.DataFrame(data)
df.head()
```

```python
df.isnull().sum()  # No missing value data set is clean
```

```python
df['Engagement_Rate'] = (
    (df['Likes'] + df['Comments'] + df['Shares']) / df['Impressions']
) * 100

df.head()
```

```python
plt.figure()
sns.barplot(x='Platform', y='Engagement_Rate', data=df)
plt.title('Average Engagement Rate by Platform')
plt.show()
```

```python
plt.figure()
sns.lineplot(x='Posting_Hour', y='Engagement_Rate', data=df)
plt.title('Engagement Rate vs Posting Time')
plt.xlabel('Posting Hour')
plt.ylabel('Engagement Rate (%)')
plt.show()
```

```python
df[['Likes','Comments','Shares']].mean().plot(kind='bar')  #ENGAGEMENT METRICS COMPARISON
plt.title('Average Likes, Comments & Shares')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,10))

# -------- KPI CARDS --------
plt.subplot(2,2,1)
plt.axis('off')

plt.text(0.1, 0.8, f"Total Likes: {df['Likes'].sum()}", fontsize=14)
plt.text(0.1, 0.6, f"Total Comments: {df['Comments'].sum()}", fontsize=14)
plt.text(0.1, 0.4, f"Total Shares: {df['Shares'].sum()}", fontsize=14)
plt.text(0.1, 0.2, f"Avg Engagement Rate: {df['Engagement_Rate'].mean():.2f}%", fontsize=14)

# -------- PLATFORM ENGAGEMENT --------
plt.subplot(2,2,2)
sns.barplot(x='Platform', y='Engagement_Rate', data=df)
plt.title('Engagement Rate by Platform')

# -------- POSTING TIME --------
plt.subplot(2,2,3)
sns.lineplot(x='Posting_Hour', y='Engagement_Rate', data=df)
plt.title('Engagement vs Posting Time')

# -------- ENGAGEMENT METRICS --------
plt.subplot(2,2,4)
df[['Likes','Comments','Shares']].mean().plot(kind='bar')
plt.title('Average Engagement Metrics')

plt.tight_layout()
plt.show()
```
